```python
""" Heaps --------------------------------------------------------------"""

import heapq

minHeap = []

heapq.heappush(minHeap,2)
heapq.heappush(minHeap,4)

print(minHeap[0])

while len(minHeap):
    print(heapq.heappop(minHeap))

# No max heap by default
#  work around is to use min heap and multiply by -1 when push and pop

maxHeap= []

heapq.heappush(maxHeap, -3)
heapq.heappush(maxHeap,-5)

print(-1 * maxHeap[0])

while len(maxHeap):
    print(-1 * heapq.heappop(maxHeap))

arr = [ 2,1,3,44,45]

heapq.heapify(arr)

while arr:
    print(heapq.heappop(arr))
```

```python
""" Nested Functions --------------------------------------------------------------"""

def outer(a,b):
    c = "c"

    def inner():
        return a+b+c
    return inner()

print(outer("a","b"))


# Can modify objects but not reassign
# unless using nonlocal keyword

def double(arr,val):
    def helper():
        for i, n in enumerate(arr):
            arr[i] *= 2

        #  will only modify val in the helper scope if nonlocal
        # val *= 2
        nonlocal val
        val *= 2
    helper()
    print(arr,val)

nums = [1,2]
val = 3

double(nums,val)
```
