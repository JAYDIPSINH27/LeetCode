### Contains Duplicate LC 217

Given an integer array nums, return true if any value appears more than once in the array, otherwise return false.

Example 1:

Input: nums = [1, 2, 3, 3]

Output: true

Example 2:

Input: nums = [1, 2, 3, 4]

Output: false


Recommended Time & Space Complexity
You should aim for a solution with O(n) time and O(n) space, where n is the size of the input array.

Intution
- Brute Force
Loop around the whole array by two loop and  keep one temp variable as counter for particular variable

Second Approach to optimize first one is:
sort the array and then check first with other point and check the counter.

Other Approach:
- HashSet
Insert element by reducing space complexity, add it to set and while adding the duplicate its available so return that there is duplicate

- HashSet Length
just compare hashset length with array and return true or false

High Level Logic

- Brute Force
Loop 1 for i in arr:
counter = 0
    Loop 2 to compare with each i in arr:
        counter += 1

    if counter > 2 return True
Return False

Time: - O(n2) , Space: O(1)

- Second Approach



Time:- O(nlogn) because sorting and swap checks, Space:- O(1) same array sorted



```python
#  Brute Force O(n2) O(1)
class Solution:

    def hasDuplicate(self, nums: List[int]) -> bool:
        for i in range(len(nums)):
            for j in range(i+1, len(nums)):
                if nums[i] == nums[j]:
                    return True
        return False                           
```

```python
#  Sorting O(nlogn) O(1) or O(n)
class Solution:

    def hasDuplicate(self, nums: List[int]) -> bool:
        nums.sort()

        for i in range(1, len(nums)):
            if nums[i] == nums[i-1]:
                return True
        return False

```

```python
##  Hash Set O(n) O(n)
class Solution:

    def hasDuplicate(self, nums: List[int]) -> bool:
        seen = set()

        for num in nums:
            if num in seen:
                return True
            seen.add(num)
        return False
```

```python
#  Hash Set Length O(n) O(n)
class Solution:

    def hasDuplicate(self, nums: List[int]) -> bool:
        return len(set(nums)) < len(nums)
```
